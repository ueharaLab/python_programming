# 配列の連結と分割

**7. 下表を作るために以下のようなプログラムを書きたい。しかしこのプログラムを実行すると下表のようにならない。誤りを見つけ出して修正せよ**


````python
import pandas as pd

L = ['Lucy',160,52,'A','USA',20,'student']
J =  ['Jordan',210,89,'O','USA',30,'NBA']
S = ['Shohei',193,80,'AB','Japan',30,'MLB']
M = ['Mami',180,62,'B','Japan',28,'wife']

profiles = L + J + S + M
profiles_df = pd.DataFrame(profiles,columns=['name','height','weight','b type','nation','age','occupation'])
print(profiles_df)
````

**8. 以下のプログラムは、上表に行を連結するプログラムである。コードを読んで構文を理解すること。次に、Rui_dfを追加するとエラーになる。原因を考えて修正せよ**


````python
profiles2 = [['Naomi',185,70,'A','Japan',30,'tennis'],
            ['Betts',177,70,'O','USA',33,'MLB']]
profiles2_df = pd.DataFrame(profiles2, columns=['name','height','weight','b type','nation','age','occupation'])

profiles_df = pd.concat([profiles_df,profiles2_df], axis =0)
print(profiles_df)
````

````python
rui = ['Rui',210,100,'O','Japan','15','NBA']
rui_df = pd.DataFrame(rui,columns=['name','height','weight','b type','nation','age','occupation'])
profiles_df = pd.concat([profiles_df,rui_df], axis =0)
print(profiles_df)
````

# 配列からの条件抽出と計算
**13 以下のプログラム例を参考にして、職業が MLB のデータをprofiles_dfから抽出せよ**

**14. profiles_df から下の計算を行うプログラムの構文を理解すること。これらを参考に、国籍がJapanのデータを取り出し、身長と体重の平均を計算せよ**
-  身長　体重の平均を計算する    
-  身長が最大の人を見つける
-  MLBの選手の平均身長を求める
-  MLBの選手の数を求める



````python

means = profiles_df['height'].mean(axis=0)
print('平均身長',means)

h_max = profiles_df['height'].max(axis=0)
print('最大身長',h_max)


mlb = profiles_df[profiles_df['occupation']=='MLB']
print('MLB選手',mlb)
h_mean = mlb['height'].mean(axis=0)
print('MLB平均身長',h_mean)
count_mlb = len(mlb)
print('MLB人数',count_mlb)


````
