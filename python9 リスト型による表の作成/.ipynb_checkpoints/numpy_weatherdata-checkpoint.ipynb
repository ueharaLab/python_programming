### numpyの活用例1

```python
#データの可視化（縦軸：気象データ、横軸：日付）
#気温
plt.plot(day, temperature)
plt.show()
#相対湿度
plt.plot(day, humidity)
plt.show()
#降水量
plt.bar(day, rainfall, align="center")
plt.show()
```

**問題1**  
2024年8月29日に観測された気温、湿度、天気、降水量を確認したい。<br>
インデックスを指定することで、各気象データの配列から2024年8月29日のデータを抽出し、出力してみよう

ヒント：インデックスは０から始まりますので、インデックスが0の場所に2024年8月1日のデータが配置されています。

```python
print( temperature[28] )
print( humidity[28] )
print( weather[28] )
print( rainfall[28] )
```

**問題2**  
2024年8月の月平均気温（8月1日～31日のすべての日の気温の平均）を計算し、出力してみよう。

ヒント：numpyには、平均を計算する便利な機能がありましたよね。

```python
print(temperature.mean())
```

**問題3**  
2024年8月29日は、この年の8月の平均と比べて、どの程度、暑かったか、または寒かったかを調べたい。<br>
2024年8月29日の気温と、2024年8月の月平均気温との差を計算し、出力することでこのことについて調べてみよう。

ヒント：難しく考えず、単純に引き算をすることで、差を計算しましょう。

```python
print(temperature[28] - temperature.mean())
```

**問題4**  
2024年8月の上旬（1～10日）と下旬（21～31日）との間では、どの程度、気温に差があっただろうか？<br>
上旬、下旬の平均気温および両者の差を計算することで、このことについて調べてみよう。

ヒント：インデックス指定による要素の抽出と平均の計算を組み合わせることで、上旬と下旬の平均気温を計算できます。

```python
#上旬の平均気温
print(temperature[0:10].mean())
#下旬の平均気温
print(temperature[21:32].mean())
#上旬と下旬の平均気温の差
print(temperature[0:10].mean() - temperature[21:32].mean())
```

**問題5**

```python
print(rainfall.sum())
```

**問題10**  
世界的に見ても、日本の夏は『蒸し暑い』ことで有名である。<br>
蒸し暑さを示す指標のひとつに『不快指数』と呼ばれるものがある。<br>
『不快指数』は気温と相対湿度を用いて、以下の計算式によって求めることができる。<br>
$$
    不快指数 = 0.81×気温 + 0.01×相対湿度×(0.99×気温ー14.3)+46.3
$$
この式に従って、2024年8月の日ごとの不快指数（discomfort_index）を計算してみよう。

ヒント：ブロードキャスト機能を使うことで、効率的に気温と相対湿度の配列（temperatureとhumidity）から不快指数が計算できます。

```python
discomfort_index = 0.81*temperature+0.01*humidity*(0.99*temperature-14.3)+46.3
print(discomfort_index)
```
