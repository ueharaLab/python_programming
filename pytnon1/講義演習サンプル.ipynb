# 演算
### 1. 整数の四則演算

```python
a = 1 + 3
print(a)
b = 3*4
print(b)
c = 5/4
print(c)
d= 5//4
print(d)
e = 5%3
print(e)
```

## 演習1. 上記の演算を以下のように実数にして試してみよ
右辺の数値のどちらか一方でも実数ならば、演算結果は実数になる

### 2. べき乗

```python
a=5**2
print(a)
b=5**(-3)
print(b)
c=3**(1/2)
print(c)
d=c**2
print(d)
```
